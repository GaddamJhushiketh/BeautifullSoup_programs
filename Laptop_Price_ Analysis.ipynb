```python
import pandas as pd
```

```python
#for reading the csv file
data = pd.read_csv("laptopData.csv")
data.head()
```

```python
#for finding the null valued row in the column
data_no_price = data[data['Price'].isna()]
data_no_price
```

```python
len(data_no_price)
```

```python
#for deletinh the null valued rows
data.dropna(subset=['Price'], inplace=True)
data
```

```python
#for removing the columns 
data.drop(columns=['Unnamed: 0'], inplace=True)
data
```

```python
#for geting pirtucular ro
for i in data:
    print(data[i][0])
```

```python
#for finding the datatypes of the columns
data.dtypes
```

```python
Q1 = data['Price'].quantile(0.25)
Q2 = data['Price'].quantile(0.75)
```

```python
#for finding the higher and lower outline of the laptop prices
IQR= Q2-Q1
lower_outline=Q1-1.5 *IQR
higher_outline=Q2+1.5 *IQR
print(lower_outline)
print(higher_outline)
```

```python
#for finding the number of laptops where expencive 
data_premium=data[data['Price']>higher_outline]
data_premium
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.histplot(data_premium['Company'])
plt.show()
```
