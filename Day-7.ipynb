```python
gdpsoup=BeautifulSoup(gdpresponse_text, "html.parser")
```

```python
rows = [i.text.strip() for i in row_tags]
rows[3:]
```

```python
gdprows = rows[3:]
gdprows
```

```python
gdpdf = pd.DataFrame(gdpdict)
gdpdf.index = range(1, len(gdpdf)+1)
gdpdf
```
