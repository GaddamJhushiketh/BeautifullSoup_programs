```python
from bs4 import BeautifulSoup
import requests
import matplotlib.pyplot as plt
import pandas as pd
```

```python
gdp=requests.get("https://en.wikipedia.org/wiki/List_of_Asian_countries_by_GDP")
gdp_text=gdp.text
type(gdp_text)
```

```python
gdpsoup = BeautifulSoup(gdp_text,"html.parser")
```

```python
gdptables = gdpsoup.find_all("table")
len(gdptables)
```

```python
heading_tag = gdptables[0].find_all("th")
heading_tag
```

```python
gdpheading = [i.text.strip() for i in heading_tag]
gdpheading
```

```python
row_tags=gdptables[0].find_all("td")
row_tags
```

```python
len(row_tags)
```

```python
rows =[i.text.strip() for i in row_tags]
rows[3:]
```

```python
gdprows = rows[3:]
gdprows
```

```python
gdpdic ={}
n=0
for i in gdpheading:
    gdpdic[i] =[gdprows[j] for j in range(n, len(gdprows), len(gdpheading))]
    n+=1
gdpdic
```

```python
gdpdf =pd.DataFrame(gdpdic)
gdpdf.index = range(1, len(gdpdf)+1)
gdpdf
```
